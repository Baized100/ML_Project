```python

# Using sklearn to perform linear regression
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from pandas.core.common import random_state
from sklearn.linear_model import LinearRegression
```

```python
# read salary dataset
salary = pd.read_csv("salary_data.csv")
salary.head(3)
```

```python
#anylyze data
salary.describe()
```

```python
# distributed visually using Seaborn distplot
# Relationship between Salary and Experience
plt.scatter(salary['YearsExperience'], salary['Salary'], color = 'green')
plt.title('Salary vs Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# Split the dataset into dependent/independent variables
# Splitting variables
x = salary.iloc[:, :1] 
y = salary.iloc[:, 1:] 
```

```python
# Splitting dataset into test/train
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python

# Train the regression model
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Prediction result
y_pred_test = regressor.predict(X_test)     # predicted value of y_test
y_pred_train = regressor.predict(X_train)
```

```python
# Prediction on training set
plt.scatter(X_train, y_train, color = 'green')
plt.plot(X_train, y_pred_train, color = 'blue')
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Sal/Exp', loc='best', facecolor='white')
plt.show()
```

```python

# Prediction on test set
plt.scatter(X_test, y_test, color = 'lightcoral')
plt.plot(X_train, y_pred_train, color = 'firebrick')
plt.title('Salary vs Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Sal/Exp', loc='best', facecolor='white')
plt.box(False)
plt.show()
```

```python
# Regressor coefficients and intercept
print(f'Coefficient: {regressor.coef_}')
print(f'Intercept: {regressor.intercept_}')
```

```python
mean_x = np.mean(x)
mean_y = np.mean(y)
slope=np.sum((x-mean_x)*(y-mean_y)) / np.sum((x-mean_x)**2)
slope
```

```python
#intercept
intercept=mean_y-slope*mean_x
intercept
```

```python
# predict value 
predicted_y = slope*2 + intercept
predicted_y
```
